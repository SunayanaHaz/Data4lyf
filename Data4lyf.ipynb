```python
import pandas as pd

df = pd.read_excel("Access_to_Tech_Dataset.xlsx")

df["domain_category"] = df["domain_category"].replace({
    "TechnologyScienceResearch": "Technology Science and Research",
    "Ecommerce": "E-commerce"
})

df = df[df["scrape_status"] == "scraped"]

df['domain_category'].value_counts()
```

```python
df['domain_category'].value_counts()
```

```python
df['domain_category'].value_counts()
```

```python
domain_counts = df["domain_category"].value_counts().reset_index()
domain_counts.columns = ["domain", "violation_count"]
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(domain_counts["domain"], domain_counts["violation_count"])
plt.xticks(rotation=45, ha="right")
plt.title("Where Technology Fails Most: Accessibility Violations by Domain")
plt.xlabel("Domain")
plt.ylabel("Number of Violations")
plt.tight_layout()
plt.show()
```

```python
top_violations = df["violation_name"].value_counts().head(10).reset_index()
top_violations.columns = ["violation", "count"]
```

```python
plt.figure()
plt.barh(top_violations["violation"], top_violations["count"])
plt.title("The Most Common Barriers to Digital Access")
plt.xlabel("Number of Violations")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
impact_counts = df.pivot_table(
    index="web_URL_id",
    columns="violation_impact",
    values="id",
    aggfunc="count",
    fill_value=0
).reset_index()
```

```python
for col in ["critical", "serious", "moderate", "minor"]:
    if col not in impact_counts.columns:
        impact_counts[col] = 0
```

```python
impact_counts["risk_score"] = (
    5*impact_counts["critical"] +
    4*impact_counts["serious"] +
    3*impact_counts["moderate"] +
    2*impact_counts["minor"]
)
```
